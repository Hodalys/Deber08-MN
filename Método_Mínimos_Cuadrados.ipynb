```python
# Datos
x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de mínimos cuadrados para un polinomio de grado 1 (línea recta)
n = len(x)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_xy = np.sum(x * y)
sum_x2 = np.sum(x**2)

a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
b = (sum_y - a * sum_x) / n

# Ecuación de la recta de mínimos cuadrados
y_aprox = a * x + b
# Calculamos el error cuadrático medio (ECM)
ecm = np.mean((y - y_aprox)**2)


# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Datos originales', color='blue')
plt.plot(x, y_aprox, label='Polinomio de grado 1', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de mínimos cuadrados (grado 1)')
plt.legend()
plt.grid(True)

# Mostramos el error
print(f'Coeficiente a1: {a}')
print(f'Coeficiente a0: {b}')
print(f'Error cuadrático medio (ECM): {ecm}')

plt.show()
```

```python
# Método de mínimos cuadrados para un polinomio de grado 2
A = np.vstack([x**2, x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[:2]

# Polinomio de grado 2
y_aprox = m[0]*x**2 + m[1]*x + m[2]

# Error cuadrático medio
ecm = np.mean((y - y_aprox)**2)

# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Datos originales', color='blue')
plt.plot(x, y_aprox, label='Polinomio de grado 2', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de mínimos cuadrados (grado 2)')
plt.legend()
plt.grid(True)

# Imprimir coeficientes y error
print(f'Coeficiente a2: {m[0]}')
print(f'Coeficiente a1: {m[1]}')
print(f'Coeficiente a0: {m[2]}')
print(f'Error cuadrático medio (ECM): {ecm}')

plt.show()
```

```python
# Método de mínimos cuadrados para un polinomio de grado 3
A = np.vstack([x**3, x**2, x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[:2]

# Polinomio de grado 3
y_aprox = m[0]*x**3 + m[1]*x**2 + m[2]*x + m[3]

# Error cuadrático medio
ecm = np.mean((y - y_aprox)**2)

# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Datos originales', color='blue')
plt.plot(x, y_aprox, label='Polinomio de grado 3', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de mínimos cuadrados (grado 3)')
plt.legend()
plt.grid(True)

# Imprimir coeficientes y error
print(f'Coeficiente a3: {m[0]}')
print(f'Coeficiente a2: {m[1]}')
print(f'Coeficiente a1: {m[2]}')
print(f'Coeficiente a0: {m[3]}')
print(f'Error cuadrático medio (ECM): {ecm}')

plt.show()
```

```python
# Función exponencial
def func(x, a, b):
    return b * np.exp(a * x)
```

```python
from scipy.optimize import curve_fit

# Ajuste de la curva
popt, pcov = curve_fit(func, x, y, p0=[1, 1])

# Coeficientes
a, b = popt
# Valores ajustados
y_fit = func(x, a, b)
# Error cuadrático medio
ecm = np.mean((y - y_fit)**2)

# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Datos originales', color='blue')
plt.plot(x, y_fit, label='Ajuste exponencial', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste exponencial por mínimos cuadrados')
plt.legend()
plt.grid(True)

# Imprimir coeficientes y error
print(f'Coeficiente a: {a}')
print(f'Coeficiente b: {b}')
print(f'Error cuadrático medio (ECM): {ecm}')
plt.show()
```

```python
# Función para ajustar (bx^a)
def func(x, a, b):
    return b * (x**a)
```

```python
# Ajuste de curva
popt, pcov = curve_fit(func, x, y, p0=[1, 1]) # p0 son valores iniciales

# Obtener los coeficientes ajustados
a_opt, b_opt = popt

# Calcular y ajustados
y_fit = func(x, a_opt, b_opt)

# Error cuadrático medio
ecm = np.mean((y - y_fit)**2)

# Imprimir los coeficientes
print(f"Coeficiente 'a': {a_opt}")
print(f"Coeficiente 'b': {b_opt}")
print(f"Error cuadrático medio (ECM): {ecm}")

# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(x, y, label='Datos originales', color='blue')
plt.plot(x, y_fit, label='Ajuste bx^a', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste por mínimos cuadrados (bx^a)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de puntuaciones de estudiantes y promedio de puntos
puntuaciones = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 27, 29, 21, 28, 28, 26, 30, 24])
promedio_puntos = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03, 3.75, 3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81])

# Método de mínimos cuadrados
n = len(puntuaciones)
sum_x = np.sum(puntuaciones)
sum_y = np.sum(promedio_puntos)
sum_xy = np.sum(puntuaciones * promedio_puntos)
sum_x2 = np.sum(puntuaciones**2)

a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
b = (sum_y - a * sum_x) / n

# Ecuación de la recta de mínimos cuadrados
y_aprox = a * puntuaciones + b

# Error cuadrático medio (ECM)
ecm = np.mean((promedio_puntos - y_aprox)**2)

# Graficar
plt.figure(figsize=(6, 4))
plt.scatter(puntuaciones, promedio_puntos, label='Datos originales', color='blue')
plt.plot(puntuaciones, y_aprox, label='Regresión lineal', color='red')
plt.xlabel('Puntuaciones de estudiantes')
plt.ylabel('Promedio de puntos')
plt.title('Ajuste de mínimos cuadrados')
plt.legend()
plt.grid(True)

# Mostrar la ecuación de la recta y el ECM
print(f"Ecuación de la recta: y = {a:.4f}x + {b:.4f}")
print(f"Error cuadrático medio (ECM): {ecm:.4f}")
plt.show()
```
